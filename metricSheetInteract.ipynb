```python
# Loop through each row of the DataFrame
def update_df(df):
    for index, row in df.iterrows():
        ticker = row["TRUE"]
        if pd.notna(ticker):  # Ensure the ticker is not NaN
            try:
                data = fetch_financial_data(ticker, session)
            except Exception as e:
                print(f"{ticker} ERROR: {e}")
                continue
            for key, value in data.items():
                if key in df.columns:
                    df.at[index, key] = value
    return df
```

```python
from googleapiclient.discovery import build
from googleapiclient.errors import HttpError


# Function to update Google Sheet with DataFrame
def update_sheet(service, spreadsheet_id, range_name, df):
    # Convert DataFrame to list of lists
    values = [df.columns.tolist()] + df.values.tolist()

    body = {"values": values}

    try:
        result = (
            service.spreadsheets()
            .values()
            .update(
                spreadsheetId=spreadsheet_id,
                range=range_name,
                valueInputOption="RAW",
                body=body,
            )
            .execute()
        )
        print(f"{result.get('updatedCells')} cells updated.")
    except HttpError as err:
        print(err)


# Update the Google Sheet
update_sheet(sheets_service, METRIC_SHEET_ID, "testing123", whole_updated_df)
```
